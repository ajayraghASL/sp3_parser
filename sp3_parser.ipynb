```python
import georinex as gr
import numpy as np
```

```python
file = "data/orbit/igu21976_00.sp3"
nav = gr.load(file)
nav
```

```python
sv = nav['sv']
sv = np.asarray(sv)
svid = []
for i in sv:
    char = str(i)
    if (char[0] == 'G'):
        svid.append(char)
print(svid)
```

```python
for i in svid:
    data = nav.sel(sv=str(i))
    time = data['time']
    time = np.asarray(time)
    position = data['position']
    position = np.asarray(position)
    for j in range(time.shape[0]):
        check = str(time[j])
        with open('data/output/orbit_{}.txt'.format(i), 'a') as f:
            if(int(check[8:10]) > 18):
                f.writelines(str(time[j])+","+str(position[j,0])+","+str(position[j,1])+","+str(position[j,2])+"\n")
```

##### Output text file format: datetime,X,Y,Z
###### 1.datetime written in datetime64[ns] format
###### 2. X,Y,Z coordinates in ECEF
